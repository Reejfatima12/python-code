```python


def plot_total_population_trend(df):
    """
    ploting 1970 to 2020 population trend in line graph
    Parameters:
        df (DataFrame): Input the population of different years in dataframe
    """
    # sum of the population with years
    trend = df.iloc[:, 5:13].sum().sort_index()
    
    # ploting line graph
    plt.plot(trend.index, trend.values, marker='o', linestyle='-')
    plt.xticks(rotation=90)
    # add the title and labels
    plt.xlabel('Year')
    plt.ylabel('Total Population')
    plt.title('Total population trend for (1970 => 2020)')
    
    # display the graph
    plt.grid(True)  # Add the grid
    plt.show()
    return

# Calling the function
plot_total_population_trend(df)
```

```python

def plot_correlation_heatmap(df):
    """
    Plotting a correlation among numeric variables
    
    Parameters:
        df (DataFrame): dataframe input have variables
    """
    sns.set(rc={"figure.figsize": (10, 10)})
    
    # Selecting numeric variables
    numeric_only = df[['Area (km²)', 'Density (per km²)', 'Growth Rate', '2022 Population']]
    
    # Calculating the correlation matrix
    correlation = numeric_only.corr()
    
    # plotting heatmao graph
    correlation_graph = sns.heatmap(correlation, annot=True)
    
    # title
    correlation_graph.set_title('Correlation Heatmap of Selected Numeric Variables')
    
    # display graph
    plt.show()
    return

# calling the defined function
plot_correlation_heatmap(df)
    
    
```

```python



def plot_continent_population_comparison(df):
    """Plot comparison of populations for different years by continent in a multi-category bar graph.

    Parameters:
        df (DataFrame): Input DataFrame containing continent-wise and year-wise population data.
    """
    # Group by continent and calculate the sum of populations for selected years
    grouped_data = df.groupby('Continent')[['2022 Population', '2015 Population', '2010 Population']].sum()
   
    # Plot the multi-category bar chart
    ax = grouped_data.plot(kind='bar', figsize=(12, 6), width=0.6)
   
    # Adding labels and title
    ax.set_xlabel('Continent')
    ax.set_ylabel('Population')
    ax.set_title('Population Comparison by Continent for 2022, 2015, and 2010')
    ax.legend(title='Year')
   
    # Show plot
    plt.show()

plot_continent_population_comparison(df)
```

```python
# skewness and kurtosis
from scipy.stats import skew, kurtosis
# Select numerical columns
numerical_columns = df.select_dtypes(include=[int, float])
# Calculating kurtosis and skewness
skewness = numerical_columns.apply(skew)
kurtosis = numerical_columns.apply(kurtosis)
# Print the results
print("Skewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis)
```
